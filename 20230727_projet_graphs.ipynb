Graphe 1
---
Réaliser un graphique avec deux axes. L'axe des abscisses va de 0 à 50 et s'intitule "Axe des x". L'axe des ordonnées s'intitule "Axe des y".
Tracer la droite d'équation : x = 3y. Avec x allant de 0 à 50.
Le graphe possède le titre suivant "Tracer une droite" et la droite est de couleur rouge.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots()

x = np.linspace(0,50)
y = x / 3

plt.plot(x, y, 'r')

plt.xlabel('Axe des x')
plt.ylabel('Axe des y')
plt.title('Tracer une droite')

plt.grid()
plt.show()
```

Graphe 7
---
Nous souhaitons un graphe qui donne l'évolution du CA pour chaque catégorie (en fonction du temps).

```python
transactions_with_products = products_data.merge(transactions_data, on='id_prod')
display(transactions_with_products)
```

```python
#mauvais exercice mais peut-être intéressant

transactions_with_products['date'] = pd.to_datetime(transactions_with_products['date'])
start_date = np.datetime64('2021-03')
dates_array = np.array([start_date + np.timedelta64(i, 'M') for i in range(12)])

fig, axes = plt.subplots()

#Cat 0
cat_0 = transactions_with_products.loc[transactions_with_products['categ'] == 0]
product_0_grouped_month = cat_0['categ'].groupby(cat_0['date'].dt.to_period('M')).value_counts().values

plt.plot(dates_array, product_0_grouped_month, label = 'Catégorie A', marker='o')

#Cat 1
cat_1 = transactions_with_products.loc[transactions_with_products['categ'] == 1]
product_1_grouped_month = cat_1['categ'].groupby(cat_1['date'].dt.to_period('M')).value_counts().values

plt.plot(dates_array, product_1_grouped_month, label = 'Catégorie A', marker='o')

#Cat 2
cat_2 = transactions_with_products.loc[transactions_with_products['categ'] == 2]
product_2_grouped_month = cat_2['categ'].groupby(cat_2['date'].dt.to_period('M')).value_counts().values

plt.plot(dates_array, product_2_grouped_month, label = 'Catégorie A', marker='o')

axes.set_ylim(0,)
plt.xlabel('Mois de l\'année')
plt.ylabel('CA de chaque catégorie')
plt.title('Evolution du CA de chaque produit par catégorie au cours de l\'année.')

plt.legend()
plt.grid()
plt.show()
```
